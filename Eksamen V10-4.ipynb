To ladde kuler, begge med masse $m = 2,0 \textnormal{ g}$ og ladning $q=3,0\cdot 10^{-7} \textnormal{ C}$ holdes med en vertikal avstand 1,0 cm fra hverandre, se figur 1. Den nederste kula holdes i ro, mens den øverste kula slippes.


Figur 1

Hva er den maksimale vertikale avstanden vi kan få mellom kulene?

---

Det er to krefter som virker på den øverste kula (hvis vi ser bort fra luftmotstand), elektrisk kraft $\vec{F_e}$ og tyngdekraft $\vec{G}$.

Siden kulene har ladning med samme fortegn er den elektriske kraften frastøtende, og $\vec{F_e}$ peker loddrett oppover.
Tyngdekraften virker loddrett nedover.

Forutsatt at $F_e > G$ i startposisjonen vil kula akselerere oppover. Når avstanden mellom de to kulene blir større, blir $F_e$ mindre, siden den er gitt ved Coulombs lov
$ F_e = k_e \frac{q\cdot q}{r^2} $.

Etterhvert vil da $ G > F_e$ og kula vil begynne å bremse, og til slutt snu.

---

Siden akselerasjonen ikke er konstant egner situasjonen seg til simulering med Eulermetoden.
Jeg velger positiv retning oppover og legger origo i sentrum av den nederste kula.

Først definerer jeg alle størrelsene som skal brukes:

````python
ke = 8.99E9
q = 3E-7
r = 0.01
m = 2.0E-3
g = 9.81
v = 0
t = 0
dt = 0.00001
````

Bevegelsen tar svært kort tid, derfor velger jeg en såpass liten verdi for for tidsintervallet ```dt```.

Med positiv retning oppover vil det høyeste punktet i bevegelsen være det punktet hvor farten ```v``` skifter fortegn fra positiv (oppover) til negativ (nedover). Derfor blir betingelsen i løkka som styrer Eulermetoden ``` while v >= 0: ```

Den elektriske kraften virker i positiv retning og tyngdekraften i negativ retning.\
Summen av kreftene blir $F_e - G$ og Newtons 2. lov gir da: $ \sum F = ma \Rightarrow a = \frac{\sum F}{m}$

Implementeringen av Eulermetoden blir da slik:

````python
while v >= 0:
    F = ke*q*q/r**2 - m*g
    a = F/m
    v = v + a*dt
    r = r + v*dt
    t = t + dt
````

Til slutt skriver jeg ut sluttresultatet, både distansen kula beveger seg før den snur, og hvor lang tid bevegelsen tar.

````python
print(f"Kula beveger seg til en avstand {r:.3} meter på {t:.3} sekunder")
````

Oppgaven spør ikke om tiden, men å vite hvor lang tid bevegelsen tar gjør det lettere å velge en fornuftig verdi for ```dt```. Når vi ser at hele bevegelsen tar under ett sekund er det naturlig å sette ```dt``` betydelig lavere enn f.eks. en bevegelse av en planet i bane rundt sola, der én runde kan ta flere år. Denne simuleringen gir riktig svar til 3 gjeldende siffer dersom $dt < 0,0001$.
